```python
Credit_card_csv = pd.read_csv('/kaggle/input/creditcardfraud/creditcard.csv')
```

```python
Credit_card_csv.head()
```

```python
Credit_card_csv.tail()
```

```python
# dataset information
Credit_card_csv.info()
```

```python
# checking the number of missing values in each column
Credit_card_csv.isnull().sum()
```

```python
# distribution of normal(legit) transaction & fraudulent transaction

#0--> Normal Transaction
#1__> fraudulent transaction

Credit_card_csv['Class'].value_counts()
```

```python
# separating the data for analysis
legit = Credit_card_csv[Credit_card_csv.Class == 0]
fraud = Credit_card_csv[Credit_card_csv.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
# compare the values for both transaction
Credit_card_csv.groupby('Class').mean()
```

```python
legit_sample = legit.sample(n=492)
```

```python
new_dataset = pd.concat([legit_sample, fraud], axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
#splitting the data into Features & Targets

X = new_dataset.drop(columns='Class', axis=1)
Y = new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

```python
#split the data into Training data & Testing Data

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
#Model Training
#Logistic Regression

model =LogisticRegression()
```

```python
# training the Logistic Regression Model with Training Data
model.fit(X_train, Y_train)
```

```python
# accuracy on training data 
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
confusion_matrix(X_train_prediction, Y_train, labels = [0,1])
```

```python
print('Accuracy on Training data : ', training_data_accuracy)
```

```python
# accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
confusion_matrix(X_test_prediction, Y_test, labels = [0, 1])
```

```python
print('Accuracy score on Test Data : ', test_data_accuracy)
```

```python
print(X_test_prediction)
```
